```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL de la página web con los datos
url = 'http://201.251.63.225/meteorologia/boletin/boletin202409w.html' ##'http://201.251.63.225/meteorologia/vp2s1/lluviasxdia.html'

# Obtener el contenido HTML de la página
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Buscar la etiqueta <pre>
pre_tag = soup.find('pre')

# Verificar si existe la etiqueta <pre>
if pre_tag:
    # Dentro de <pre>, buscar el primer <hr>
    hr_tag = pre_tag.find('hr')
    
    # Si se encuentra el <hr>, obtener el texto que sigue después de <hr>
    if hr_tag and hr_tag.next_sibling:
        pre_text = hr_tag.next_sibling.get_text(strip=True)
    else:
        raise ValueError("No se encontró el contenido después de la etiqueta <hr> en <pre>.")
else:
    raise ValueError("No se encontró la etiqueta <pre> en la página web.")

# Procesar el texto en filas y columnas
rows = pre_text.strip().split('\n')
data = [row.split() for row in rows]

# Crear un DataFrame de Pandas con los datos extraídos
df = pd.DataFrame(data[1:], columns=data[0])
##print(df)
# Convertir las columnas numéricas de texto a números (si aplica)
for col in df.columns:
    df[col] = df[col] # saco el controlador de tipo numerico pd.to_numeric()

# Mostrar el DataFrame resultante
print(df)
```
